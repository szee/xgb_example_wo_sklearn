```python
import pandas as pd
import numpy as np
from scipy import stats

from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix

import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df_x = pd.read_csv('x_train.csv', header=None, delimiter=';')
df_y = pd.read_csv('y_train.csv', header=None)
```

```python
np_x = np.loadtxt('x_train.csv', delimiter=';')
np_y = np.loadtxt('y_train.csv')
np_x_test = np.loadtxt('x_test.csv', delimiter=';')
```

```python
np_x_sq = np.exp(np_x)
np_x_test_sq = np.exp(np_x_test)
```

```python
stats.describe(np_x)
```

```python
model = XGBClassifier()
#model.fit(np_x, np_y)
#model
```

```python
dtrain = xgb.DMatrix(np_x_sq, label=np_y)
dtest = xgb.DMatrix(np_x_test_sq)
```

```python
# setup parameters for xgboost
param = {}
# use softmax multi-class classification
param['objective'] = 'multi:softmax'

param['learning_rate'] = 0.3
param['n_estomators'] = 7000
param['eta'] = 0.01
param['max_depth'] = 3
param['silent'] = 1
param['nthread'] = 4
param['num_class'] = 5
param['gamma'] = 0.3
param['reg_alpha'] = 0.3
param['seed'] = 27


num_round = 100
```

```python
xgb.cv(param, dtrain, nfold=6, metrics={'merror'})
```

```python
model.fit(param, dtrain, num_round)
# get prediction
pred = model.predict(dtest)
print(model.feature_importances_)
```
